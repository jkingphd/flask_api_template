```python
from sklearn.datasets import load_iris
import pandas as pd
import requests
import json

url = 'http://localhost:8000/display/'
```

Load example data.

```python
data = load_iris()
X = data['data']
columns = data['feature_names']
df = pd.DataFrame(X, columns=columns)
df.head()
```

Send data to flask app.

```python
url = 'http://localhost:8000/predict/'
response = requests.post(url, data=json.dumps({'input':df.to_json()}), headers={'Content-Type':'application/json'})
result = pd.read_json(response.json()['output'])
result.head()
```
